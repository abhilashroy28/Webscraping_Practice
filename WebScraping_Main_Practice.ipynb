```python
import requests
import pandas as pd 
import numpy as np
from bs4 import BeautifulSoup
import re
```

```python
url = "https://keithgalli.github.io/web-scraping/webpage.html"
```

```python
r = requests.get(url)
```

```python
html_content = r.content
```

```python
soup = BeautifulSoup(html_content, "html.parser")
```

```python
print(soup.prettify())
```

## Exercise #1: Grab all social links on webpage in 3 different ways

#### In 3 different ways

```python
links = soup.select("ul.socials a")
actual_links = [link["href"] for link in links]
actual_links
```

```python
ulink = soup.find("ul", attrs={"class": "socials"})
links = ulink.find_all("a")
actual_links = [link["href"] for link in links]
actual_links
```

```python
links = soup.select("li.social a")
actual_links = [link["href"] for link in links]
actual_links
```

## Exercise #2: Scrape an HTML table into a Pandas Dataframe

```python
table = soup.select("table.hockey-stats")[0]
columns = table.find("thead").find_all("th")
column_names = [c.string for c in columns]
column_names
table_rows = table.find("tbody").find_all("tr")
l = []  
for tr in table_rows:
    td = tr.find_all('td')
    row = [str(tr.get_text()).strip() for tr in td]    # get_text() [is for nested strings] or string or text
    l.append(row)
df = pd.DataFrame(l, columns=column_names)
df
```
